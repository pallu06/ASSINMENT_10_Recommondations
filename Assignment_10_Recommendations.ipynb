# EDA

```python
book1 = book.rename({'User.ID' : 'userid','Book.Title' : 'title', 'Book.Rating' : 'rating'} , axis  = 1)
```

```python
book1
```

```python
book2 = book1.iloc[:,1:]
book2
```

```python
#number of unique users in the dataset
len(book2.userid.unique())
```

```python
len(book2.title.unique())
```

```python
len(book2.rating.unique())
```

```python
book3 = book2.drop_duplicates()
book3
```

```python
user_book = book3.pivot_table(index='userid',
                                 columns='title',
                                 values='rating').reset_index(drop=True)
```

```python
user_book
```

```python
user_book.index = book3.userid.unique()
user_book
```

```python
#Impute those NaNs with 0 values
user_book.fillna(0, inplace=True)
```

```python
user_book
```

# Finding Cosine Similarities score

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book3.userid.unique()
user_sim_df.columns = book3.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
